Visualizations for Insights

```python
quarterly_columns = [col for col in final_cqv.columns if "quarter" in col]
yearly_columns = [str(year) for year in range(2014, 2021)]

# Convert to long format for quarterly data
quarterly_data = final_cqv.melt(id_vars=["Country"], 
                           value_vars=quarterly_columns, 
                           var_name="Quarter", 
                           value_name="Visitors")

# Convert to long format for yearly data
yearly_data = final_cqv.melt(id_vars=["Country"], 
                        value_vars=yearly_columns, 
                        var_name="Year", 
                        value_name="Visitors")

# Convert the "Year" column to numeric for sorting
yearly_data["Year"] = yearly_data["Year"].astype(int)

# Aggregate yearly data for all countries
total_yearly_visitors = yearly_data.groupby("Year")["Visitors"].sum().reset_index()

# Plot total yearly visitors
plt.figure(figsize=(12, 6))
sns.lineplot(data=total_yearly_visitors, x="Year", y="Visitors", marker="o", color="blue")
plt.title("Total Yearly Visitors (2014-2020)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Visitors", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Aggregate yearly data for the top 5 countries with the highest total visitors
top_countries = yearly_data.groupby("Country")["Visitors"].sum().nlargest(5).index
top_countries_data = yearly_data[yearly_data["Country"].isin(top_countries)]

# Plot yearly trends for the top 5 countries
plt.figure(figsize=(14, 8))
sns.lineplot(data=top_countries_data, x="Year", y="Visitors", hue="Country", marker="o")
plt.title("Yearly Visitor Trends for Top 5 Countries (2014-2020)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Visitors", fontsize=12)
plt.legend(title="Country")
plt.grid(True)
plt.show()

# Ensure the Quarter column is treated as a string
top_10_data["Quarter"] = top_10_data["Quarter"].astype(str)


# Extract year information and convert it to an integer
top_10_data["Quarter"] = top_10_data["Quarter"].str.extract(r"(\d{4})").astype(int)

# Summarize quarterly visitors for the top 10 countries
total_quarterly_visitors = top_10_data.groupby("Quarter")["Visitors"].sum().reset_index()

# Plot quarterly trends for the top 10 countries
plt.figure(figsize=(14, 8))
sns.lineplot(data=top_10_data, x="Quarter", y="Visitors", hue="Country", alpha=0.6)
plt.title("Quarterly Visitor Patterns for Top 10 Countries", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Quarterly Visitors", fontsize=12)
plt.grid(True)
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
quarters = final_cqv.melt(
    id_vars=["Country"],
    value_vars=[col for col in final_cqv.columns if "quarter" in col],
    var_name="Quarter",
    value_name="Value"
)

# Extract year and quarter from the Quarter column
quarters['Year'] = quarters['Quarter'].str.extract(r'(\d{4})').astype(int)
quarters['Quarter'] = quarters['Quarter'].str.extract(r'(1st|2nd|3rd|4th)').replace({
    "1st": "Q1",
    "2nd": "Q2",
    "3rd": "Q3",
    "4th": "Q4"
})

# Plotting data for a few countries to showcase trends
selected_countries = ['Canada', 'Argentina', 'Brazil', 'Mexico', 'Austria']
filtered_data = quarters[quarters['Country'].isin(selected_countries)]

# Line plot for quarter-wise trends
plt.figure(figsize=(14, 8))
sns.lineplot(
    data=filtered_data,
    x="Year",
    y="Value",
    hue="Country",
    style="Quarter",
    markers=True
)
plt.title("Quarter-wise Trends for Selected Countries", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.legend(title="Country / Quarter")
plt.grid(visible=True, linestyle='--', alpha=0.6)
plt.savefig("quarterwisetrends.png")
plt.show()
```
