```python
cag_updated = pd.merge(cag, cyv, on='Country', how='inner')

# Display the first few rows of the merged DataFrame
cag_updated.head()
```

```python
years = ['2014', '2015', '2016', '2017', '2018', '2019', '2020']

# Create a copy of the original data for manipulation
new_cag = cag_updated.copy()

# Loop through each year and multiply quarterly percentages by the yearly total
for year in years:
    # Define quarterly columns for the current year
    age_grp = [f'{year} 0-14', f'{year} 15-24',
                f'{year} 25-34', f'{year} 35-44',f'{year} 45-54', f'{year} 55-64', f'{year} 65 AND ABOVE']

    existing_age_grp = [col for col in age_grp if col in cag_updated.columns]
    

    for age_range in existing_age_grp:
        new_cag[age_grp] = cag_updated[age_grp] * cag_updated[year] / 100

# Display the updated DataFrame with calculated values
new_cag.head()
   
```

```python
cag_updated.to_csv('cag_updated.csv', index = False)
```

```python
visitor_columns = [col for col in cag_updated.columns if col.isdigit()]  # Total visitors columns
percentage_columns = [col for col in cag_updated.columns if not col.isdigit() and any(year in col for year in visitor_columns)]

# Create a copy of the DataFrame to modify percentages into absolute values
cag_processed = cag_updated.copy()

# Convert percentage columns into absolute values based on corresponding total visitors
for year in visitor_columns:
    year_percentage_cols = [col for col in percentage_columns if col.startswith(year)]
    if year_percentage_cols:
        for col in year_percentage_cols:
            cag_processed[col] = (cag_processed[col] / 100) * cag_processed[year]

# Display the updated DataFrame
cag_processed.head()
```

```python
# Cleaning column names for easier processing
cag_processed.columns = cag_processed.columns.str.strip().str.replace(' ', '_').str.replace('AND', 'AND_')

# Reshaping the data to long format for easier plotting
long_data = pd.melt(cag_processed, id_vars=['Country'], var_name='Year_AgeGroup', value_name='Population')

# Splitting 'Year_AgeGroup' into 'Year' and 'AgeGroup' for clarity
long_data[['Year', 'AgeGroup']] = long_data['Year_AgeGroup'].str.split('_', n=1, expand=True)
long_data['Year'] = long_data['Year'].astype(int)  # Converting Year to integer

# Dropping the original 'Year_AgeGroup' column
long_data.drop(columns=['Year_AgeGroup'], inplace=True)

# Aggregating population by year for visualizations
aggregated_data = long_data.groupby(['Year', 'AgeGroup']).Population.sum().reset_index()

# Creating visualizations
plt.figure(figsize=(18, 6))

# Visualization 1: Line chart for population trends by age group
sns.lineplot(data=aggregated_data, x='Year', y='Population', hue='AgeGroup', marker='o')
plt.title('Population Trends by Age Group Over Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualization 2: Bar chart for a specific year's population by age group
year_data = aggregated_data[aggregated_data['Year'] == 2020]  # Select the year 2020
plt.figure(figsize=(12, 6))
sns.barplot(data=year_data, x='AgeGroup', y='Population', palette='viridis')
plt.title('Population Distribution by Age Group (2020)', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```

```python
# Calculate total population for each country across all years and age groups
total_population_by_country = long_data.groupby('Country').Population.sum().sort_values(ascending=False)

# Select the top 5 countries by total population
top_5_countries = total_population_by_country.head(5).index.tolist()

# Filter the data for the top 5 countries
top_5_data = long_data[long_data['Country'].isin(top_5_countries)]

# Aggregate population by country and age group
top_5_agegroup_data = top_5_data.groupby(['Country', 'AgeGroup']).Population.sum().reset_index()

# Create a grouped bar chart to show relationships between age groups for top 5 countries
plt.figure(figsize=(14, 8))
sns.barplot(data=top_5_agegroup_data, x='AgeGroup', y='Population', hue='Country', palette='Set2')
plt.title('Population Distribution by Age Group for Top 5 Countries', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("population_age.png")
plt.show()
```
